```python
import numpy as np
import matplotlib.pyplot as plt
```

----
## 1D line

We first explore the **fractal dimension** of a straight line by covering it with boxes of decreasing size.

```python
x = np.array([0,1])
y = np.zeros_like(x)

fig,ax = plt.subplots(1,1,figsize=(10,2))
ax.set_axis_off()
ax.plot(x,y,lw=10,color='grey')
```

We use `1`, `2`, `3`, and `4` boxes to cover the line.

```python
smax = 5
fig,axs = plt.subplots(smax-1,1,figsize=(10,5))
for s in range(1,smax):


    xbox = np.linspace(x.min(),x.max(),s+1)
    axs[s-1].set_axis_off()
    axs[s-1].plot(x,y,lw=10,color='grey')
    axs[s-1].plot(xbox,-0.02*np.ones(s+1),lw=2,color='red')
    axs[s-1].plot(xbox,+0.02*np.ones(s+1),lw=2,color='red')
    for i in range(xbox.shape[0]):
        axs[s-1].plot([xbox[i],xbox[i]],[-0.02,+0.02],lw=2,color='red')
    axs[s-1].text(0.9,0.012,'s='+str(s))
    
```

We observe:
- The discretisation $s$ always doubles.
- The number of boxes $N(s)$ containing the line increases with $s$

|s|N(s)|
|-|----|
|1|1|
|2|2|
|3|3|
|4|4|

Plotting $s$ versus $N(s)$:
- once on a **linear** plot
- a second time on a **log-log** plot

```python
s = np.array([1,2,3,4,5])
N = s

fig,axs = plt.subplots(1,2,figsize=(10,4))
axs[0].set_title('Linear')
axs[0].set_xlabel('$s$')
axs[0].set_ylabel('$N(s)$')
axs[0].plot(s,N)

axs[1].set_title('Log-Log')
axs[1].set_xlabel('$\log s$')
axs[1].set_ylabel('$\log N(s)$')
axs[1].set_xscale('log')
axs[1].set_yscale('log')
axs[1].plot(s,N)
plt.tight_layout()
```

From the log-log plot, we can deduce a relation between discretisation $s$ and number of counted boxes $N$:
$$
N(s) \sim s^{D}
$$
Here, $D$ is the **fractal dimension**. 

In the case of the straight line, we find by linear regression `D=1`.

```python
D = np.polyfit(np.log10(s),np.log10(N),1)
print('D: ',D)
```

----
## 2D box

We observe:
- The discretisation $s$ always doubles.
- The number of boxes $N(s)$ containing the box increases with the square of $s$

|s|N(s)|
|-|----|
|1|1|
|2|4|
|3|9|
|4|16|

Plotting $s$ versus $N(s)$:
- once on a **linear** plot
- a second time on a **log-log** plot

```python
s = np.array([1,2,3,4,5,6,7])
N = s**2

fig,axs = plt.subplots(1,2,figsize=(10,4))
axs[0].set_title('Linear')
axs[0].set_xlabel('$s$')
axs[0].set_ylabel('$N(s)$')
axs[0].plot(s,N)

axs[1].set_title('Log-Log')
axs[1].set_xlabel('$\log s$')
axs[1].set_ylabel('$\log N(s)$')
axs[1].set_xscale('log')
axs[1].set_yscale('log')
axs[1].plot(s,N)
plt.tight_layout()
```

From the log-log plot, we can deduce a relation between discretisation $s$ and number of counted boxes $N$,
$n \sim s^D$:
$$
N(s) \sim \big( \frac{1}{s} \big)^{-D}
$$
Here, $D$ is the **fractal dimension**. 

In the case of the straight line, we find by linear regression `D=2`.

```python
D = np.polyfit(np.log10(s),np.log10(N),1)
print('D: ',D)
```
