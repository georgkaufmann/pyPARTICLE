```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

[see](https://github.com/isellsoap/deutschlandGeoJSON)

```python
f = open('data/Border_Germany.json')

data = json.load(f)
coord = data['features'][0]['geometry']['coordinates']
print(len(coord))
for i in range(len(coord)):
    for j in range(len(coord[i])):
        
        #print([coord[i][j][:][0]])
        for k in range(len(coord[i][j])):
            #print(i,j,k,coord[i][j][k][0],coord[i][j][k][1])
            #print(coord[i][j][k][0])
            plt.plot(coord[i][j][k][0],coord[i][j][k][1],lw=1,marker='o',markersize=2)
            
```
