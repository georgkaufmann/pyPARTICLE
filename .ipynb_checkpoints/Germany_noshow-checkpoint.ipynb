```python
f = open('data/Border_Germany.json')

data = json.load(f)
coord = data['features'][0]['geometry']['coordinates']
print(len(coord))
for i in range(len(coord)):
    for j in range(len(coord[i])):
        
        #print([coord[i][j][:][0]])
        for k in range(1):#len(coord[i][j])):
            #print(i,j,k,coord[i][j][k][0],coord[i][j][k][1])
            #print(coord[i][j][k][0])
            plt.plot(coord[i][j][k][0],coord[i][j][k][1],lw=1,marker='o',markersize=2)
            plt.fill(coord[i][j][k][0],coord[i][j][k][1])
            
```

```python
name = 'data/germany-detailed-boundary_917.geojson'
f = open(name)
data = json.load(f)
npoly = len(data['features'][0]['geometry']['coordinates'])
print('npoly: ',npoly)

for ipoly in range(npoly):
    coord = data['features'][0]['geometry']['coordinates'][ipoly][0]

    x,y = np.zeros(len(coord)), np.zeros(len(coord))
    for i in range(len(coord)):
        #print(coord[i])
        x[i] = coord[i][0]
        y[i] = coord[i][1]
    plt.fill(x,y)
```
