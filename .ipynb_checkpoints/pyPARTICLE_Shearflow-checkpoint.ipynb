# `pyPARTICLE`

pyPARTICLE, a program package for particle flow and transport.

```python
import numpy as np
import matplotlib.pyplot as plt
import libPARTICLE
```

----
## Horizontal shear flow
In this notebook, we implement a more realistic flow component for **shear flow**.
As driving forces, we introduce:
- **Couette flow:** Induced by a shear along the top of the fluid layer, with horizontal velocity $v_x$.
- **Poiseuille flow:** Driven by a horizontal pressure gradient with pressure $p_x$.

<img src='images/couette_and_pipe_flow.jpg' style=height:50mm;>

The **horizontal velocity** of particles in the fluid is then given by:
$$\fbox{$
 v_x(z) = \frac{1}{2\eta} \frac{dp}{dx} \left( z^2 - z h \right) + v_0 \left( 1 - \frac{z}{h} \right)
$}$$
with
- $z$ [m] depth coordinate
- $h$ [m] channel height
- $v_x$ [m/s] horizontal velocity profile
- $v_0$ [m/s] horizontal velocity along top
- $\frac{dp}{dx}$ [Pa/m] - horizontal pressure gradient
- $\eta$ [Pas ] dynamic viscosity

```python
def particleVelocityShear(z,h,dpdx=0.,v0=0,TC=10):
    """
    Particle motion: 
    Couette and/or Poiseuille shear-flow profile
    input:
      z        - vertical coordninate
      h        - fluid layer thickness
      dpdx     - pressure gradient
      v0       - horizontal velocity along surface 
    output:
      vShear   - horizontal shear velocity profile
    """
    vShear = 1/2/libPARTICLE.water_viscosity(TC) * dpdx * ( z**2 - z*h ) + v0*(1. - z/h)
    return vShear
```

```python
h = 1.
z = np.linspace(0,h,101)

plt.figure(figsize=(10,6))
plt.title('Shear velocity')
plt.xlabel('Shear velocity $v_s$ [m/s]')
plt.ylabel('Depth $z$ [m]')
plt.ylim([1,0])
plt.plot(particleVelocityShear(z,h,v0=0.2),z,lw=4,label='Couette')
plt.plot(particleVelocityShear(z,h,dpdx=-0.005),z,lw=4,label='Poiseuille')
plt.plot(particleVelocityShear(z,h,v0=0.2,dpdx=-0.005),z,lw=4,label='Channel')
plt.legend()
plt.grid()
```

## Example 1: Shear flow
Run an example with a pure Couette-flow profile...

```python
Nfloat=1000;Nsettled=100
sidex=0.5;sidey=0.2
seedx=0.15;seedy=0.12
sigmax=0.05;sigmay=0.01
xdiff=0.005;ydiff=0.01
xadv=1.0;yadv=-0.00
path='div/shear1';name='shear1'

pX,pY,pState,pMaterial = libPARTICLE.particleInit(Nfloat=Nfloat,Nsettled=Nsettled,sidex=sidex,sidey=sidey,seedx=seedx,seedy=seedy,sigmax=sigmax,sigmay=sigmay)
materials = libPARTICLE.particleMaterials(show=False)
libPARTICLE.particlePlot(pX,pY,pState,pMaterial,sidex=sidex,sidey=sidey,path=path,name=name)
libPARTICLE.particleRun(pX,pY,pState,pMaterial,materials,tmax=0.40,dt=0.01,sidex=sidex,sidey=sidey,xdiff=xdiff,ydiff=ydiff,xadv=xadv,yadv=yadv,path=path,name=name)
```

With the shear velocity of $v_x=1$ m/s along the top, and running the model for 0.4 s, we can nicely see
the particle cloud being faster on top the the bottom part.
